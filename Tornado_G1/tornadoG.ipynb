```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Data/us_tornado_dataset_1950_2021.csv')
df
```

```python
new_col_names= {'yr':'Year', 'mo':'Month', 'dy':'Day', 'st':'State', 'mag':'EFS','date':'Date','len':'Distance', 'wid':'width'}
clean_df= df.rename(columns=new_col_names)
clean_df
```

```python
clean_df.info()
```

```python
clean_df['Date'] = pd.to_datetime(clean_df['Date'])
print(clean_df.info())
```

```python
tefs_freq=clean_df['EFS'].value_counts()
plt.figure(figsize=(10, 6))
tefs_freq.plot(kind='bar')
plt.xlabel('EFS Tornado Type')
plt.ylabel('Frequency')
plt.title('Frequency of Each EFS Tornado Type')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
dist_df=clean_df.sort_values(by='Distance', ascending=False)
dist_df
```

```python
EF5_df= clean_df[clean_df['EFS'] == 5]
EF5_df
```

```python
ef5_freq= EF5_df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(ef5_freq.index, ef5_freq.values)
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Frequency of EF5 Occurrences per Year')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
start_date = '2011-01-01'
end_date = '2021-12-31'
years_df= clean_df[(clean_df['Date'] >= start_date) & (clean_df['Date']<= end_date)]
years_df
```

```python
t_freq= clean_df['State'].value_counts()
plt.figure(figsize=(10, 6))
t_freq.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.title('Frequency of Tornado by State 1950 to 2021')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
