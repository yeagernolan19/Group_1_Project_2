```python
import pandas as pd
import sqlite3
```

```python
df= pd.read_csv('Data/us_tornado_dataset_1950_2021.csv')
df
```

```python
df.info()
```

```python
cols_to_keep= ['date','st', 'mag', 'slat', 'slon', 'elat', 'elon']
clean_cols_df= df[cols_to_keep]
clean_cols_df
```

```python
new_col_names= {'st':'State', 'mag':'EFS','date':'Date'}
clean_df= clean_cols_df.rename(columns=new_col_names)
clean_df
```

```python
clean_df['Date'] = pd.to_datetime(clean_df['Date'])
print(clean_df.info())
```

```python
start_date = '2011-01-01'
end_date = '2021-12-31'
years_df= clean_df[(clean_df['Date'] >= start_date) & (clean_df['Date']<= end_date)]
years_df
```

```python
years_df.reset_index(drop=True, inplace=True)
years_df['Tornado_id']=None
for index, row in years_df.iterrows():
    tornado_id=f"T{index + 1}"
    years_df.loc[index, 'Tornado_id']= tornado_id
years_df
```

```python
years_df = years_df[['Tornado_id'] + [col for col in years_df.columns if col != 'Tornado_id']]
years_df.reset_index(drop=True, inplace=True)
years_df
```

```python
EFS_3_df = years_df[years_df['EFS'] == 3].copy()
EFS_3_df
```

```python
EFS_4_df = years_df[years_df['EFS'] == 4].copy()
EFS_4_df
```

```python
EFS_5_df = years_df[years_df['EFS'] == 5].copy()
EFS_5_df
```

```python
con=sqlite3.connect('Tornado.db')
cur=con.cursor()
years_df.to_sql('2011_to_2021_Tornadoes', con, index=False)
EFS_3_df.to_sql('EFS_3_Tornadoes', con, index=False)
EFS_4_df.to_sql('EFS_4_Tornadoes', con, index=False)
EFS_5_df.to_sql('EFS_5_Tornadoes', con, index=False)
```
